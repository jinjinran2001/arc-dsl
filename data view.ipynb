```python
import random
from typing import Tuple, List
import numpy as np

def generate_grid(squares, size=10):
    """
    Generates a grid with 4s placed at the corners of specified squares.

    Parameters:
    - squares: a list of tuples, each representing a square defined by
               (x1, y1, x2, y2) where (x1, y1) is the top-left corner and
               (x2, y2) is the bottom-right corner.
    - size: the size of the grid (default is 10 for a 10x10 grid).

    Returns:
    - A tuple of tuples representing the grid.
    """
    # Initialize the grid with zeros
    grid = [[0]*size for _ in range(size)]
    
    for square in squares:
        x1, y1, x2, y2 = square  # Unpack square coordinates
        # Define the corners of the square
        corners = [(x1, y1), (x1, y2), (x2, y1), (x2, y2)]
        for x, y in corners:
            # Check if the coordinates are within the grid bounds
            if 0 <= x < size and 0 <= y < size:
                grid[x][y] = 4  # Place a 4 at the corner position
    # Convert the grid to a tuple of tuples before returning
    return tuple(tuple(row) for row in grid)

def print_grid(grid: Tuple[Tuple[int, ...], ...]):
    """Helper function to print the grid in a readable format"""
    for row in grid:
        print(row)

# Example usage:
if __name__ == "__main__":
    # Generate a grid with 2 squares
    grid = generate_square_grid(num_squares=2)
    print("Generated grid:")
    print_grid(grid)
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, Normalize

def plot_task(
    task: List[dict],
    title: str = None
) -> None:
    """
    displays a task
    """
    cmap = ListedColormap([
        '#000', '#0074D9', '#FF4136', '#2ECC40', '#FFDC00',
        '#AAAAAA', '#F012BE', '#FF851B', '#7FDBFF', '#870C25'
    ])
    norm = Normalize(vmin=0, vmax=9)
    args = {'cmap': cmap, 'norm': norm}
    height = 2
    width = len(task)
    figure_size = (width * 3, height * 3)
    figure, axes = plt.subplots(height, width, figsize=figure_size)
    if width > 1:
        for column, example in enumerate(task):
            axes[0, column].imshow(example['input'], **args)
            axes[1, column].imshow(example['output'], **args)
            axes[0, column].axis('off')
            axes[1, column].axis('off')
    else:
        example=task[0]
        axes[0].imshow(example['input'], **args)
        axes[1].imshow(example['output'], **args)
        axes[0].axis('off')
        axes[1].axis('off')
    if title is not None:
        figure.suptitle(title, fontsize=20)
    plt.subplots_adjust(wspace=0.1, hspace=0.1)
    plt.show()
```

```python
def transform_grid(input_grid, transformation):
    """
    Transforms the input grid based on the specified transformation.

    Parameters:
    - input_grid: The original grid to be transformed.
    - transformation: A function that defines how to transform each square.

    Returns:
    - A new grid after applying the transformation.
    """
    size = len(input_grid)
    # Find all positions with '4's (corners of squares)
    squares = []
    visited = set()
    for x in range(size):
        for y in range(size):
            if input_grid[x][y] == 4 and (x, y) not in visited:
                # Find the opposite corner of the square
                for dx in range(x+1, size):
                    for dy in range(y+1, size):
                        if input_grid[dx][dy] == 4:
                            if input_grid[x][dy] == 4 and input_grid[dx][y] == 4:
                                # Found a square from (x, y) to (dx, dy)
                                squares.append((x, y, dx, dy))
                                visited.update([(x, y), (dx, dy), (x, dy), (dx, y)])
                                break
                    else:
                        continue
                    break

    # Apply the transformation to each square
    transformed_squares = [transformation(square) for square in squares]

    # Generate the new grid with transformed squares
    new_grid = generate_grid(transformed_squares, size)
    return new_grid
```

```python
# Define the transformation function
def expand_and_shift(square):
    # print(square)
    x1, y1, x2, y2 = square
    # Expand the square size by increasing the distance between corners
    # and shift it towards the top-left corner
    new_x1 = max(0, x1 - 2)
    new_y1 = max(0, y1 - 2)
    new_x2 = min(9, x2 + 1)
    new_y2 = min(9, y2 + 1)
    return (new_x1, new_y1, new_x2, new_y2)

# Original grid (First input grid)
grid1 = generate_grid(squares=[(3, 3, 5, 5)], size=10)

# Transform the grid
grid2 = transform_grid(grid1, expand_and_shift)
```

```python
def generate_grid(size, elements):
    """
    Generates a grid with specified elements placed at given positions.
    """
    rows, cols = size
    grid = [[0 for _ in range(cols)] for _ in range(rows)]
    
    for elem in elements:
        number = elem.get('number', 0)
        
        if 'positions' in elem:
            # Place the number at specified positions
            for pos in elem['positions']:
                r, c = pos
                if 0 <= r < rows and 0 <= c < cols:
                    grid[r][c] = number
                    
        elif 'shape' in elem and 'start_pos' in elem:
            shape = elem['shape']
            r0, c0 = elem['start_pos']
            transform = elem.get('transform', lambda s: s)
            shape = transform(shape)
            
            for i, row in enumerate(shape):
                for j, val in enumerate(row):
                    if val != 0:
                        r, c = r0 + i, c0 + j
                        if 0 <= r < rows and 0 <= c < cols:
                            grid[r][c] = val
                        
    return tuple(tuple(row) for row in grid)

def rotate_shape(shape, angle):
    """
    Rotates the shape by the given angle (90, 180, 270 degrees).
    """
    if angle == 90:
        return [list(row)[::-1] for row in zip(*shape)]
    elif angle == 180:
        return [row[::-1] for row in shape[::-1]]
    elif angle == 270:
        return [list(row) for row in zip(*shape)][::-1]
    else:
        return shape

def reflect_shape(shape, axis):
    """
    Reflects the shape across the specified axis ('horizontal' or 'vertical').
    """
    if axis == 'horizontal':
        return shape[::-1]
    elif axis == 'vertical':
        return [row[::-1] for row in shape]
    else:
        return shape

# First Input Grid
size1 = (12, 13)
positions_5_1 = [(1, 4), (6, 1), (6, 11), (11, 4)]
shape1 = [
    [1, 1, 1],
    [1, 0, 1],
    [1, 1, 1]
]
elements1 = [
    {'number': 5, 'positions': positions_5_1},
    {'shape': shape1, 'start_pos': (4, 5)}
]
grid1 = generate_grid(size1, elements1)

# Second Input Grid
size2 = (15, 13)
positions_5_2 = [(1, 5), (5, 1), (8, 10), (12, 5)]
shape2 = [
    [3, 3, 3],
    [3, 0, 3],
    [3, 3, 3]
]
elements2 = [
    {'number': 5, 'positions': positions_5_2},
    {'shape': shape2, 'start_pos': (5, 4)}
]
grid2 = generate_grid(size2, elements2)

# Third Input Grid
size3 = (15, 14)
positions_5_3 = [(2, 6), (6, 12), (8, 1), (13, 7)]
shape3 = [
    [4, 4, 4, 4],
    [4, 0, 0, 4],
    [4, 0, 0, 4],
    [4, 4, 4, 4]
]
elements3 = [
    {'number': 5, 'positions': positions_5_3},
    {'shape': shape3, 'start_pos': (5, 4)}
]
grid3 = generate_grid(size3, elements3)

# Optional: Rotate the shape in the third grid
shape3_rotated = rotate_shape(shape3, 90)
elements3_rotated = [
    {'number': 5, 'positions': positions_5_3},
    {'shape': shape3_rotated, 'start_pos': (5, 4)}
]
grid3_rotated = generate_grid(size3, elements3_rotated)
```

```python
def generate_grid(size, clusters):
    """
    Generates a grid and places clusters into it.

    Parameters:
    - size: tuple (rows, cols) defining the grid size.
    - clusters: list of dictionaries, each specifying:
        - 'numbers': 2D list of numbers in the cluster.
        - 'position': (row, col) where to place the cluster.
        - 'transform': function to transform the cluster (optional).

    Returns:
    - A tuple of tuples representing the grid.
    """
    rows, cols = size
    grid = [[0 for _ in range(cols)] for _ in range(rows)]

    for cluster in clusters:
        numbers = cluster['numbers']
        position = cluster['position']
        transform = cluster.get('transform', lambda x: x)
        numbers = transform(numbers)
        grid = place_cluster(grid, numbers, position)

    return tuple(tuple(row) for row in grid)

def place_cluster(grid, numbers, position):
    """
    Places a cluster of numbers into the grid at the specified position.

    Parameters:
    - grid: The grid to place numbers into.
    - numbers: 2D list of numbers representing the cluster.
    - position: (row, col) where to place the cluster.

    Returns:
    - The updated grid.
    """
    rows = len(grid)
    cols = len(grid[0])
    cluster_rows = len(numbers)
    cluster_cols = len(numbers[0])
    start_row, start_col = position

    for i in range(cluster_rows):
        for j in range(cluster_cols):
            r = start_row + i
            c = start_col + j
            if 0 <= r < rows and 0 <= c < cols:
                if numbers[i][j] != 0:
                    grid[r][c] = numbers[i][j]
    return grid

def rotate_cluster(cluster_numbers, angle):
    """
    Rotates the cluster by the specified angle (90, 180, 270 degrees).

    Parameters:
    - cluster_numbers: 2D list of numbers in the cluster.
    - angle: The angle to rotate (90, 180, 270).

    Returns:
    - The rotated cluster numbers.
    """
    if angle == 90:
        return [list(row) for row in zip(*cluster_numbers[::-1])]
    elif angle == 180:
        return [row[::-1] for row in cluster_numbers[::-1]]
    elif angle == 270:
        return [list(row) for row in zip(*cluster_numbers)][::-1]
    else:
        return cluster_numbers  # No rotation

def mirror_cluster(cluster_numbers, axis):
    """
    Mirrors the cluster across the specified axis.

    Parameters:
    - cluster_numbers: 2D list of numbers in the cluster.
    - axis: 'horizontal' or 'vertical'.

    Returns:
    - The mirrored cluster numbers.
    """
    if axis == 'horizontal':
        return cluster_numbers[::-1]
    elif axis == 'vertical':
        return [row[::-1] for row in cluster_numbers]
    else:
        return cluster_numbers  # No mirroring
```
