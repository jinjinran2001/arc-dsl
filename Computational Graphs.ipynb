```python
import ast
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Arrow
import os

class Node:
    def __init__(self, name):
        self.name = name
        self.children = []

def parse_function_composition(composition):
    """Parse a function composition string into an AST."""
    return ast.parse(composition).body[0].value

def build_tree(node):
    """Recursively build a tree from an AST node."""
    if isinstance(node, ast.Call):
        tree_node = Node(node.func.id)
        for arg in node.args:
            child = build_tree(arg)
            if child:
                tree_node.children.append(child)
        return tree_node
    return None

def layout_tree(root, x=0, y=0, level=0, x_spacing=2, y_spacing=2):
    """Assign (x, y) coordinates to each node in the tree, from bottom to top."""
    root.x, root.y = x, y
    if not root.children:
        return x + x_spacing, y, 1
    
    total_width = 0
    max_depth = 0
    for child in root.children:
        next_x, next_y, depth = layout_tree(child, x + total_width, y + y_spacing, level + 1, x_spacing, y_spacing)
        total_width = next_x - x
        max_depth = max(max_depth, depth)
    
    root.x = x + total_width / 2 - x_spacing / 2
    return x + total_width, y, max_depth + 1

def draw_node(ax, node, radius=0.5):
    """Draw a node on the matplotlib axes with the label inside."""
    circle = Circle((node.x, node.y), radius, fill=True, facecolor='lightblue', edgecolor='black', zorder=10)
    ax.add_patch(circle)
    ax.text(node.x, node.y, node.name, ha='center', va='center', fontweight='bold', fontsize=10, zorder=11)

def draw_edge(ax, parent, child):
    """Draw an edge between parent and child nodes with arrow pointing from child to parent."""
    dx = parent.x - child.x
    dy = parent.y - child.y
    arrow = Arrow(child.x, child.y, dx, dy, width=0.3, color='gray', zorder=5)
    ax.add_patch(arrow)

def draw_tree(ax, node):
    """Recursively draw the entire tree."""
    for child in node.children:
        draw_edge(ax, node, child)
        draw_tree(ax, child)
    draw_node(ax, node)  # Draw nodes last to ensure they're on top of the edges

def visualize_computation_graph(composition, name=None, save_path=None):
    """
    Main function to visualize the computation graph of a function composition from bottom to top.
    
    Args:
    composition (str): The function composition string.
    save_path (str, optional): If provided, save the graph to this path instead of displaying it.
    """
    ast_tree = parse_function_composition(composition)
    tree = build_tree(ast_tree)
    
    fig, ax = plt.subplots(figsize=(12, 8))
    layout_tree(tree, y_spacing=2)
    draw_tree(ax, tree)
    
    ax.set_aspect('equal')
    ax.autoscale_view()
    ax.axis('off')
    if name is None:
        plt.title(f"Computation Graph: {composition}")
    else:
        plt.title(f"Computation Graph: {name}")
    plt.tight_layout()
    
    if save_path:
        os.makedirs(os.path.dirname(save_path), exist_ok=True)
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        plt.close(fig)
        print(f"Graph saved to {save_path}")
    else:
        plt.show()

# Example usage
compositions = [
    "f(g(h(x)))",
    "a(b(c(d(x), e(y))))",
    "p(q(r(s(t(x)))))"
]

for i, comp in enumerate(compositions):
    print(f"Visualizing: {comp}")
    # Display the graph
    visualize_computation_graph(comp)
    
    # Save the graph
    save_path = f"bottom_up_graph_{i+1}.png"
    # visualize_computation_graph(comp, save_path)
```

```python
import ast
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Arrow
import os
from collections import defaultdict
import colorsys

class Node:
    def __init__(self, name):
        self.name = name
        self.children = []
        self.subgraph_hash = None
        self.color = 'lightblue'  # Default color

def parse_function_composition(composition):
    """Parse a function composition string into an AST."""
    return ast.parse(composition).body[0].value

def build_tree(node):
    """Recursively build a tree from an AST node, excluding variables."""
    if isinstance(node, ast.Call):
        tree_node = Node(node.func.id)
        for arg in node.args:
            child = build_tree(arg)
            if child:
                tree_node.children.append(child)
        return tree_node
    return None  # This will filter out all non-function nodes, including variables

def compute_subgraph_hash(node):
    """Compute a hash for the subgraph rooted at this node."""
    if not node.children:
        return hash(node.name)
    child_hashes = tuple(sorted(compute_subgraph_hash(child) for child in node.children))
    return hash((node.name, child_hashes))

def identify_shared_subgraphs(root):
    """Identify the largest shared subgraphs in the tree."""
    subgraph_counts = defaultdict(list)
    
    def traverse(node):
        node.subgraph_hash = compute_subgraph_hash(node)
        subgraph_counts[node.subgraph_hash].append(node)
        for child in node.children:
            traverse(child)
    
    traverse(root)
    
    # Filter out subgraphs that aren't shared or are part of larger shared subgraphs
    shared_subgraphs = {}
    for h, nodes in subgraph_counts.items():
        if len(nodes) > 1:
            is_largest = True
            for node in nodes:
                parent = next((p for p in subgraph_counts.values() if node in p and len(p) > 1 and p != nodes), None)
                if parent:
                    is_largest = False
                    break
            if is_largest:
                shared_subgraphs[h] = nodes
    
    return shared_subgraphs

def generate_colors(n):
    """Generate n distinct colors."""
    HSV_tuples = [(x * 1.0 / n, 0.7, 0.7) for x in range(n)]
    return [colorsys.hsv_to_rgb(*x) for x in HSV_tuples]

def color_shared_subgraphs(root, shared_subgraphs):
    """Color nodes in shared subgraphs."""
    colors = generate_colors(len(shared_subgraphs))
    color_map = dict(zip(shared_subgraphs.keys(), colors))
    
    def color_traverse(node):
        if node.subgraph_hash in color_map:
            stack = [node]
            while stack:
                current = stack.pop()
                current.color = color_map[node.subgraph_hash]
                stack.extend(current.children)
        else:
            for child in node.children:
                color_traverse(child)
    
    color_traverse(root)

def layout_tree(root, x=0, y=0, level=0, x_spacing=2, y_spacing=2):
    """Assign (x, y) coordinates to each node in the tree, from bottom to top."""
    root.x, root.y = x, y
    if not root.children:
        return x + x_spacing, y, 1
    
    total_width = 0
    max_depth = 0
    for child in root.children:
        next_x, next_y, depth = layout_tree(child, x + total_width, y + y_spacing, level + 1, x_spacing, y_spacing)
        total_width = next_x - x
        max_depth = max(max_depth, depth)
    
    root.x = x + total_width / 2 - x_spacing / 2
    return x + total_width, y, max_depth + 1

def draw_node(ax, node, radius=0.5):
    """Draw a node on the matplotlib axes with the label inside."""
    circle = Circle((node.x, node.y), radius, fill=True, facecolor=node.color, edgecolor='black', zorder=10)
    ax.add_patch(circle)
    ax.text(node.x, node.y, node.name, ha='center', va='center', fontweight='bold', fontsize=10, zorder=11)

def draw_edge(ax, parent, child):
    """Draw an edge between parent and child nodes with arrow pointing from child to parent."""
    dx = parent.x - child.x
    dy = parent.y - child.y
    arrow = Arrow(child.x, child.y, dx, dy, width=0.3, color='gray', zorder=5)
    ax.add_patch(arrow)

def draw_tree(ax, node):
    """Recursively draw the entire tree."""
    for child in node.children:
        draw_edge(ax, node, child)
        draw_tree(ax, child)
    draw_node(ax, node)

def visualize_computation_graph_with_shared_subgraphs(composition, name=None, save_path=None):
    """
    Main function to visualize the computation graph with colored shared subgraphs, excluding variables.
    
    Args:
    composition (str): The function composition string.
    save_path (str, optional): If provided, save the graph to this path instead of displaying it.
    """
    ast_tree = parse_function_composition(composition)
    tree = build_tree(ast_tree)
    
    shared_subgraphs = identify_shared_subgraphs(tree)
    if len(shared_subgraphs) == 0:
        return
    color_shared_subgraphs(tree, shared_subgraphs)
    
    fig, ax = plt.subplots(figsize=(20, 12))  # Large figure size for complex graphs
    layout_tree(tree, y_spacing=2, x_spacing=3)
    draw_tree(ax, tree)
    
    ax.set_aspect('equal')
    ax.autoscale_view()
    ax.axis('off')
    if name is None:
        plt.title(f"Computation Graph with Shared Subgraphs: {composition}")
    else:
        plt.title(f"Computation Graph with Shared Subgraphs: {name}")
    plt.tight_layout()
    
    if save_path:
        os.makedirs(os.path.dirname(save_path), exist_ok=True)
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        plt.close(fig)
        print(f"Graph saved to {save_path}")
    else:
        plt.show()

# Example usage
composition = "paint(apply(lbind(apply, maximum), papply(pair, I, dmirror(I))), mapply(lbind(shift, toobject(difference(asindices(I), ofcolor(apply(lbind(apply, maximum), papply(pair, I, dmirror(I))), ZERO)), apply(lbind(apply, maximum), papply(pair, I, dmirror(I))))), pair(interval(invert(NINE), NINE, ONE), interval(NINE, invert(NINE), NEG_ONE))))"

print(f"Visualizing: {composition}")
visualize_computation_graph_with_shared_subgraphs(composition)

save_path = "function_only_shared_subgraph_graph.png"
# visualize_computation_graph_with_shared_subgraphs(composition, save_path)
```

```python
import ast
from typing import Set

def get_grid_producing_functions(dsl_content: str) -> Set[str]:
    """Extract function names that return Grid type from the DSL."""
    grid_functions = set()
    
    # Parse the DSL content
    tree = ast.parse(dsl_content)
    
    for node in ast.walk(tree):
        if isinstance(node, ast.FunctionDef):
            # Look for return type hints
            returns_annotation = node.returns if hasattr(node, 'returns') else None
            if returns_annotation:
                # Convert the annotation to a string
                return_type = ast.unparse(returns_annotation)
                # Check if the function returns a Grid type
                if 'Grid' in return_type:
                    grid_functions.add(node.name)
            
            # Also check docstring for grid-related comments
            docstring = ast.get_docstring(node)
            if docstring and any(keyword in docstring.lower() for keyword in ['grid', 'subgrid']):
                grid_functions.add(node.name)
    
    return grid_functions

def highlight_grid_nodes(tree_node, grid_functions):
    """
    Recursively process the computation tree and highlight nodes that produce valid grids.
    
    Args:
    tree_node: Node object from the computation graph
    grid_functions: Set of function names that produce Grid type outputs
    
    Returns:
    None (modifies the tree nodes in place)
    """
    # Check if current node is a grid-producing function
    if tree_node.name in grid_functions:
        tree_node.color = 'lightgreen'  # Highlight grid nodes
        
    # Recursively process children
    for child in tree_node.children:
        highlight_grid_nodes(child, grid_functions)

def analyze_and_highlight_grid_nodes(composition: str, dsl_content: str):
    """
    Main function to analyze a composition and highlight nodes that represent valid grids.
    
    Args:
    composition: The function composition string to analyze
    dsl_content: Content of the DSL implementation file
    
    Returns:
    root_node: The root node of the processed tree with highlighted grid nodes
    """
    # Parse the composition into an AST
    ast_tree = parse_function_composition(composition)
    
    # Build the computation tree
    root_node = build_tree(ast_tree)
    
    # Get the set of grid-producing functions from the DSL
    grid_functions = get_grid_producing_functions(dsl_content)
    
    # Known grid-producing functions from analysis
    grid_functions.update({
        'paint',        # paint object to grid
        'fill',         # fill value at indices
        'crop',         # subgrid specified by start and dimension
        'canvas',       # grid construction
        'rot90',        # quarter clockwise rotation
        'rot180',       # half rotation
        'rot270',       # quarter anticlockwise rotation
        'hmirror',      # mirroring along horizontal
        'vmirror',      # mirroring along vertical
        'dmirror',      # mirroring along diagonal
        'cmirror',      # mirroring along counterdiagonal
        'hconcat',      # concatenate two grids horizontally
        'vconcat',      # concatenate two grids vertically
        'upscale',      # upscale grid
        'downscale',    # downscale grid
        'underfill',    # fill value at indices that are background
        'underpaint'    # paint object to grid where there is background
    })
    
    # Highlight the nodes in the tree
    highlight_grid_nodes(root_node, grid_functions)
    
    return root_node

composition = "paint(canvas(0, (3, 3)), toobject(asindices(I)))"

# Analyze and highlight grid nodes
root = analyze_and_highlight_grid_nodes(composition, 'dsl.py')
```

```python
import ast
from typing import Dict, Any, List, Tuple, Set
import inspect
import dsl
import re

class StepEvaluator:
    def __init__(self):
        self.functions = {
            name: func for name, func in inspect.getmembers(dsl, inspect.isfunction)
            if not name.startswith('_')
        }
        self.constants = {
            'NEG_ONE': -1, 'NEG_TWO': -2, 'ZERO': 0, 'ONE': 1, 'TWO': 2,
            'THREE': 3, 'FOUR': 4, 'FIVE': 5, 'SIX': 6, 'SEVEN': 7,
            'EIGHT': 8, 'NINE': 9, 'TEN': 10, 'DOWN': (1, 0), 'RIGHT': (0, 1),
            'UP': (-1, 0), 'LEFT': (0, -1), 'ORIGIN': (0, 0),
            'UNITY': (1, 1), 'NEG_UNITY': (-1, -1), 'UP_RIGHT': (-1, 1),
            'DOWN_LEFT': (1, -1), 'ZERO_BY_TWO': (0, 2),
            'TWO_BY_ZERO': (2, 0), 'TWO_BY_TWO': (2, 2),
            'THREE_BY_THREE': (3, 3), 'T': True, 'F': False
        }
        self.temp_values: Dict[str, Any] = {}
        self.current_step_inputs: Dict[str, Any] = {}
        self.pending_steps: Dict[str, str] = {}
        
    def evaluate_name(self, node: ast.Name) -> Any:
        if node.id in self.temp_values:
            self.current_step_inputs[node.id] = self.temp_values[node.id]
            return self.temp_values[node.id]
        elif node.id == 'I':
            self.current_step_inputs['I'] = self.input_grid
            return self.input_grid
        elif node.id in self.constants:
            return self.constants[node.id]
        elif node.id in self.functions:
            return self.functions[node.id]
        elif node.id in self.pending_steps:
            # Evaluate pending step first
            pending_expr = self.pending_steps[node.id]
            result = self.evaluate_expression(pending_expr)
            self.temp_values[node.id] = result
            self.current_step_inputs[node.id] = result
            return result
        else:
            raise ValueError(f"Unknown variable or function: {node.id}")
        
    def evaluate_call(self, node: ast.Call) -> Any:
        if isinstance(node.func, ast.Name):
            func_name = node.func.id
            if func_name not in self.functions:
                raise ValueError(f"Unknown function: {func_name}")
            func = self.functions[func_name]
        else:
            raise ValueError(f"Unsupported function reference type: {type(node.func)}")
            
        args = []
        for arg in node.args:
            if isinstance(arg, ast.Name):
                args.append(self.evaluate_name(arg))
            elif isinstance(arg, ast.Call):
                args.append(self.evaluate_call(arg))
            else:
                raise ValueError(f"Unsupported argument type: {type(arg)}")
                
        return func(*args)
    
    def evaluate_expression(self, expr: str) -> Any:
        """Evaluate a single expression"""
        tree = ast.parse(expr, mode='eval')
        return self.evaluate_call(tree.body)
    
    def evaluate_step(self, step_expr: str, result_var: str) -> Tuple[Any, Dict[str, Any]]:
        """Evaluate a step and return both result and inputs used"""
        self.current_step_inputs = {}
        result = self.evaluate_expression(step_expr)
        self.temp_values[result_var] = result
        return result, dict(self.current_step_inputs)

    def evaluate_solution(self, solution, input_grid) -> List[Tuple[str, str, Any, Dict[str, Any]]]:
        """Evaluate all steps in a solution"""
        self.input_grid = input_grid
        self.temp_values.clear()
        self.pending_steps.clear()
        
        # First collect all steps
        for step in solution.steps:
            self.pending_steps[step.result_var] = step.expression
            
        step_results = []
        # Then evaluate them in order
        for step in solution.steps:
            try:
                result, inputs = self.evaluate_step(step.expression, step.result_var)
                step_results.append((step.result_var, step.expression, result, inputs))
            except Exception as e:
                print(f"Error evaluating step {step.result_var} = {step.expression}")
                print(f"Current temp_values: {list(self.temp_values.keys())}")
                raise
            
        return step_results
```

```python
import ast
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Arrow
import os
from collections import defaultdict
import colorsys

class Node:
    def __init__(self, name):
        self.name = name
        self.children = []
        self.subgraph_hash = None
        self.color = 'lightblue'  # Default color
        self.path = name  # Initialize path with node name
        self.full_expression = None  # Store the full expression to reach this node

def parse_function_composition(composition):
    """Parse a function composition string into an AST."""
    return ast.parse(composition).body[0].value

def build_tree_with_paths(node, parent_path=""):
    """Recursively build a tree from an AST node, tracking function paths."""
    if isinstance(node, ast.Call):
        tree_node = Node(node.func.id)
        
        # Build the expression for this node
        args_expressions = []
        for arg in node.args:
            child = build_tree_with_paths(arg, f"{tree_node.name}")
            if child:
                tree_node.children.append(child)
                args_expressions.append(child.full_expression)
            elif isinstance(arg, ast.Name):
                args_expressions.append(arg.id)
                
        # Construct the full expression for this node
        if args_expressions:
            tree_node.full_expression = f"{tree_node.name}({', '.join(args_expressions)})"
        else:
            tree_node.full_expression = tree_node.name
            
        return tree_node
    elif isinstance(node, ast.Name):
        tree_node = Node(node.id)
        tree_node.full_expression = node.id
        return tree_node
    return None

def compute_subgraph_hash(node):
    """Compute a hash for the subgraph rooted at this node."""
    if not node.children:
        return hash(node.name)
    child_hashes = tuple(sorted(compute_subgraph_hash(child) for child in node.children))
    return hash((node.name, child_hashes))

def identify_shared_subgraphs(root):
    """Identify the largest shared subgraphs in the tree."""
    subgraph_counts = defaultdict(list)
    
    def traverse(node):
        node.subgraph_hash = compute_subgraph_hash(node)
        subgraph_counts[node.subgraph_hash].append(node)
        for child in node.children:
            traverse(child)
    
    traverse(root)
    
    shared_subgraphs = {}
    for h, nodes in subgraph_counts.items():
        if len(nodes) > 1:
            is_largest = True
            for node in nodes:

                parent = next((p for p in subgraph_counts.values() if node in p and len(p) > 1 and p != nodes), None)

                if parent:
                    is_largest = False
                    break
            if is_largest:
                shared_subgraphs[h] = nodes

    return shared_subgraphs

def generate_colors(n):
    """Generate n distinct colors."""
    HSV_tuples = [(x * 1.0 / n, 0.7, 0.7) for x in range(n)]
    return [colorsys.hsv_to_rgb(*x) for x in HSV_tuples]

def color_shared_subgraphs(root, shared_subgraphs):
    """Color nodes in shared subgraphs."""
    colors = generate_colors(len(shared_subgraphs))
    color_map = dict(zip(shared_subgraphs.keys(), colors))
    
    def color_traverse(node):
        if node.subgraph_hash in color_map:
            stack = [node]
            while stack:
                current = stack.pop()
                current.color = color_map[node.subgraph_hash]
                stack.extend(current.children)
        else:
            for child in node.children:
                color_traverse(child)
    
    color_traverse(root)

def layout_tree(root, x=0, y=0, level=0, x_spacing=2, y_spacing=2):
    """Assign (x, y) coordinates to each node in the tree."""
    root.x, root.y = x, y
    if not root.children:
        return x + x_spacing, y, 1
    
    total_width = 0
    max_depth = 0
    for child in root.children:
        next_x, next_y, depth = layout_tree(child, x + total_width, y + y_spacing, level + 1, x_spacing, y_spacing)
        total_width = next_x - x
        max_depth = max(max_depth, depth)
    
    root.x = x + total_width / 2 - x_spacing / 2
    return x + total_width, y, max_depth + 1

def draw_node_with_path(ax, node, radius=0.5):
    """Draw a node on the matplotlib axes with the label inside and path as tooltip."""

    
    # Add the full expression as text below the node
    # ax.text(node.x, node.y - radius - 0.2, node.full_expression, 
    #         ha='center', va='top', fontsize=8, wrap=True,
    #         bbox=dict(facecolor='white', alpha=0.7, edgecolor='none'))
    output = evaluate_expression(node.full_expression, I)
    if isinstance(output, frozenset):
        output = list(output)
    if isinstance(output, list):
        node_color = "green"
    else:
        node_color = 'lightblue'
    circle = Circle((node.x, node.y), radius, fill=True, facecolor=node_color, edgecolor=node_color, zorder=10)
    ax.add_patch(circle)
    ax.text(node.x, node.y, node.name, ha='center', va='center', fontweight='bold', fontsize=10, zorder=11)
    ax.text(node.x, node.y - radius - 0.2, output, 
            ha='center', va='top', fontsize=8, wrap=True,
            bbox=dict(facecolor='white', alpha=0.7, edgecolor='none'))

def draw_edge(ax, parent, child):
    """Draw an edge between parent and child nodes."""
    dx = parent.x - child.x
    dy = parent.y - child.y
    arrow = Arrow(child.x, child.y, dx, dy, width=0.3, color='gray', zorder=5)
    ax.add_patch(arrow)

def draw_tree(ax, node):
    """Recursively draw the entire tree."""
    for child in node.children:
        draw_edge(ax, node, child)
        draw_tree(ax, child)

    draw_node_with_path(ax, node)

def visualize_computation_graph_with_paths(composition, name=None, save_path=None):
    """
    Main function to visualize the computation graph with paths to each node.
    
    Args:
    composition (str): The function composition string.
    name (str, optional): Custom name for the graph title.
    save_path (str, optional): If provided, save the graph to this path.
    """
    ast_tree = parse_function_composition(composition)
    tree = build_tree_with_paths(ast_tree)
    
    shared_subgraphs = identify_shared_subgraphs(tree)
    if shared_subgraphs:
        color_shared_subgraphs(tree, shared_subgraphs)
    
    fig, ax = plt.subplots(figsize=(24, 16))  # Larger figure size for paths
    layout_tree(tree, y_spacing=4, x_spacing=6)  # Increased spacing for paths
    draw_tree(ax, tree)
    
    ax.set_aspect('equal')
    ax.autoscale_view()
    ax.axis('off')
    
    title = f"Computation Graph with Function Paths: {name if name else composition}"
    plt.title(title, pad=20, wrap=True)
    
    plt.tight_layout()
    
    if save_path:
        os.makedirs(os.path.dirname(save_path), exist_ok=True)
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        plt.close(fig)
        print(f"Graph saved to {save_path}")
    else:
        plt.show()

# Example usage
if __name__ == "__main__":
    composition = "fill(I, THREE, delta(ofcolor(I, EIGHT)))"    
    print(f"Visualizing with paths: {composition}")
    visualize_computation_graph_with_paths(composition, "Function Composition with Paths")
```
